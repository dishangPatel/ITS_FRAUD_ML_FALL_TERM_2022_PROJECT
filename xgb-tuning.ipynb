```python
train=pd.read_csv('../input/trainnntesttt/preprocessesTrain.csv')
test=pd.read_csv('../input/trainnntesttt/preprocessedTest (2).csv')
```

```python
x=train.drop(['isFraud'],axis=1)
y=train['isFraud']
```

```python
x=(x-x.min())/(x.max()-x.min())
test=(test-test.min())/(test.max()-test.min())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test = train_test_split(x,y ,random_state=104, test_size=0.2, shuffle=True)
```

```python
from xgboost import XGBClassifier
XGBClassifier()
```

```python
model=XGBClassifier(n_estimators=1000,max_depth=25,min_child_weight=5)
model.fit(x,y)
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import roc_auc_score
print(roc_auc_score(y_test,y_pred))
```

```python
y_final=model.predict(test)
```

```python
import sklearn
hyperparameter_grid = {
    'regressor__n_estimators': [50,100, 500, 1000],
    'regressor__max_depth': [ 9, 12 , 16, 20 , 25],
    'regressor__learning_rate': [0.01, 0.03, 0.05, 0.1, 0.3]
}


random_cv = sklearn.model_selection.RandomizedSearchCV(estimator=XGBClassifier(),
            param_distributions=hyperparameter_grid,
            cv=3, 
            n_iter=5,
            scoring = 'roc_auc',
            n_jobs = -1,
            verbose = 5, 
            return_train_score = True,
            random_state=42)



random_cv.fit(X_train, y_train)
```

```python
best=random_cv.best_estimator_
```
